# 4. Checking for Mixed & Data Types (Data Types)

## airport_codes_df

```python
# check for mixed types in the airport_codes_df
for cols in airport_codes_df.columns.tolist():
    mixed_type = (airport_codes_df[[cols]].applymap(type) != airport_codes_df
[[cols]].iloc[0].apply(type)).any(axis = 1)
    if len(airport_codes_df[mixed_type])>0:
        print(cols)
# no mixed type columnes
```

## delta_df

```python
# check for mixed types in the delta_df
for cols in delta_df.columns.tolist():
    mixed_type = (delta_df[[cols]].applymap(type) != delta_df
[[cols]].iloc[0].apply(type)).any(axis = 1)
    if len(delta_df[mixed_type])>0:
        print(cols)
```
