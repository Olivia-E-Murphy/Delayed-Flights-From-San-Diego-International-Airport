# 3. Correlation Matrix heatmap

```python
# creates subset for variables suitable for the matrix
airlines_correlation = airlines_final_df[['Known Total Delay (Minutes)', 'Scheduled Elapsed Time (Minutes)', 'Actual Elapsed Time (Minutes)', 'Delay Carrier (Minutes)', 'Delay Weather (Minutes)', 'Delay National Aviation System (Minutes)',  'Delay Security (Minutes)', 'Delay Late Aircraft Arrival (Minutes)', 'Difference in Actual & Expected Airtime']]
```

```python
# checks first five rows of airlines_correlation 
airlines_correlation.head()
```

```python
# returns correlation coefficiants for airlines_correlation
airlines_correlation.corr()
```

```python
# creates heatmap for airlines_correlation with matplotlib
f = plt.figure(figsize = (12, 12))
matplotlib_corr_matrix = plt.matshow(airlines_correlation.corr(), fignum = f.number)
plt.xticks(range(airlines_correlation.shape[1]), airlines_correlation.columns, fontsize = 10, rotation = 90)
plt.yticks(range(airlines_correlation.shape[1]), airlines_correlation.columns, fontsize = 10)
plt.title('Airline Variables Correlation Matrix', fontsize = 30)
colorbar = plt.colorbar()
colorbar.ax.tick_params(labelsize = 10)
```

```python
plt.subplots(figsize = (12, 12))
sns_heatmap_correlation = sns.heatmap(airlines_correlation.corr(), annot = True)
plt.title('Airline Variables Correlation Matrix', fontsize = 20)
```

## What do the coeeficients in the plot mean?

## There is a strong relationship between the following variables: Delay Late Aircraft Arrival (Minutes) & Known Total Delay Time (Minutes), Delay Carrier (Minutes) & Known Total Delay Time (Minutes), Actual Elapsed Time (Minutes) & Scheduled Elapsed Time (Minutes)
## There is a moderate relationship between:  Known Total Delay (Minutes) & Delay National Aviation System (Minutes).

## There is a weak relationship between the following vairabes:  Known Total Delay (Minutes)  & Delay Weather (Minutes).

## All other variables have a very weak relationship

# 4. Scatterplot

```python
# returns a scatter plot for Actual Elapsed Time (Minutes) and Scheduled Elapsed Time (Minutes)
sns.lmplot (data = airlines_correlation, x = 'Actual Elapsed Time (Minutes)', y = 'Scheduled Elapsed Time (Minutes)')
plt.title('Relationship Between Actual Elapsed Time (Minutes) and Scheduled Elapsed Time (Minutes)')
```

## Interpretation: The relationship between Actual Elapsed Time (Minutes) and Scheduled Elapsed Time (Minutes) is linear for the most part (except for when the Actual Elapsed Time is 0)
## Overall, the higher the Actual Elapsed Time, the higher the scheduled elasped time.

```python
# returns scatterplot for Known Total Delay (Minutes) and Delay Carrier (Minutes)
sns.lmplot(data = airlines_correlation, x = 'Delay Carrier (Minutes)', y = 'Known Total Delay (Minutes)')
plt.title('Relationship Between Known Total Delay (Minutes) and Delay Carrier (Minutes)')
```

## Interpretation: The relationship between Known Total Delay (Minutes) and Delay Carrier (Minutes) is not linear, but there is some sort of relationship between Known Total Delay (Minutes) and Delay Carrier (Minutes).
## The delay due to Carrier is contributing to the total known delay, but there are other factors.

```python
# returns scatterplot for Delay Late Aircraft Arrival (Minutes) & Departure Delay (Minutes)
sns.lmplot(data = airlines_correlation, x = 'Delay Late Aircraft Arrival (Minutes)', y = 'Known Total Delay (Minutes)')
plt.title('Relationship Between Delay Late Aircraft Arrival (Minutes) & Departure Delay (Minutes)')
```

# The relationship between Delay Late Aircraft Arrival (Minutes) & Known Total Delay (Minutes) is not exactly linear
### The delay due to a late aircraft is contributing to the total known delay, but there are other factors
